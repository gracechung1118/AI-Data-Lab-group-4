## Hotels near Mercedes Benz Stadium


```python
import pandas as pd
from geopy.distance import geodesic

# --- 1. Define Target and Load Data ---

stadium_coords = (33.7555, -84.4008)

# Load your hotel data from a CSV
try:
    df = pd.read_csv("Hotel.csv")
except FileNotFoundError:
    print("Error: hotel.csv not found.")
    print("Make sure the file is in the same directory.")
    exit()

# --- 2. Calculate Distances Efficiently ---

df['distance_to_stadium_miles'] = df.apply(
    lambda row: geodesic(
        stadium_coords, (row['latitude'], row['longitude'])
    ).miles,
    axis=1
)

# --- 3. Sort and Display Results ---

df_sorted = df.sort_values(by="distance_to_stadium_miles")

print(f"Top 10 Closest Hotels to Mercedes-Benz Stadium:\n")
print(df_sorted.head(10))
```
