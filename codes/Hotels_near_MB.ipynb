```python
# --- 4. Sort and Display Results ---

# Sort the entire DataFrame by the new distance column
df_sorted = df.sort_values(by="distance_to_stadium_miles")

# Select only the columns we care about for the final output
# We'll also rename 't' and 's' to be more descriptive
final_columns = {
    'name': 'Hotel Name',
    't': 'Latitude',
    's': 'Longitude',
    'distance_to_stadium_miles': 'Distance (Miles)'
}

# Keep only the columns we want and rename them
df_final_output = df_sorted[final_columns.keys()].rename(columns=final_columns)

# Print the top 10 closest hotels
print(f"Top 10 Closest Hotels to Mercedes-Benz Stadium:\n")
print(df_final_output.head(10).to_string(index=False))
```
